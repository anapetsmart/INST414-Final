```python
import json
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
```

```python
breast_cancer_data = pd.DataFrame(breast_cancer['data'], columns = breast_cancer['feature_names'])
breast_cancer_data['tumor_type'] = pd.DataFrame(breast_cancer['target']).apply(lambda x: breast_cancer['target_names'][x])
```

```python
breast_cancer_data
breast_cancer_dict = breast_cancer_data.to_dict('index')
for i in breast_cancer_dict:
    symptoms = list(breast_cancer_dict[i].values())
    symptoms = symptoms[:-1]
    breast_cancer_dict[i] = ({'type':[breast_cancer_dict[i]['tumor_type']],'symptoms': set(symptoms)})
    print(breast_cancer_dict[i])
```

```python
breast_cancer_test_data = breast_cancer_data[:113]
breast_cancer_train_data = breast_cancer_data[113:]
breast_cancer_test = breast_cancer_test_data.to_dict('index')
breast_cancer_train = breast_cancer_train_data.to_dict('index')
    
```
