```python
import json
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
```

```python
breast_cancer_data = pd.DataFrame(breast_cancer['data'], columns = breast_cancer['feature_names'])
#breast_cancer_data['tumor_type'] = pd.DataFrame(breast_cancer['target']).apply(lambda x: breast_cancer['target_names'][x])
breast_cancer_data = breast_cancer_data.fillna(0)
```

```python
breast_cancer_dict = breast_cancer_data.to_dict('index')
# for i in breast_cancer_dict:
#     symptoms = list(breast_cancer_dict[i].values())
#     breast_cancer_dict[i] = ({'symptoms': set(symptoms)})
#     print(breast_cancer_dict[i])
```

```python
from scipy.sparse import lil_matrix
import scipy.spatial.distance
target_case = 0
symptoms = breast_cancer_data.loc[target_case]
distances = scipy.spatial.distance.cdist(breast_cancer_data, [symptoms], metric = 'euclidean')[:,0]
query_distances = list(zip(breast_cancer_data.index, distances))
for id, i in sorted(query_distances, key=lambda x: x[1], reverse=False)[:10]:
    patient_id = breast_cancer_dict[id]
    print(id, breast_cancer_dict[id], i)
```
